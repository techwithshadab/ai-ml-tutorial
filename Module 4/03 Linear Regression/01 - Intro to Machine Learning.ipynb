## Machine Learning

 - Quick History
     - In the 1990s, Machine Learning became popular as classical stastical principles were leveraged on growing computer power.
     - This new marriage of computer science and statistics gave birth to what we now call Data Science.
     - Scientists were now able to build intelligent systems that could analyze and learn from massive amounts of data.
 <br><br>    
 - What is Machine Learning (ML)
     - In 1959, Arthur Samuel defined ML as a, “Field of study that gives computers the ability to learn without being explicitly programmed.”
     - Basically, Data Scientists use ML algorithms that receive input data and use statistical analysis to predict an output.
         - Many ML algorithms can learn and update their outputs as new data becomes available.

## Supervised Learning: An Overview

- There are broadly 2 types of tasks in Supervised Learning — **Regression** and **Classification**

- In **Regression**, an algorithm is used to predict _continuous_ values
    - Continuous values are numbers like price, salary, age, etc. 
    - An example would be predicting monthly sales in a LS&Co store<br><br> 
- In **Classification**, an algorithm is used to predict _discrete_ values
    - Discrete values are defined categories. Examples include Yes or No, Male or Female, and Spam or not Spam. 
    - Another example would be to predict whether a LS&Co customer is loyal or not

### Ground Truth Context-Dependent

- **Ground Truth** also known as the "y", label, dependent variable are the "answers" a supervised learning algorithm is trying to predict
    - Context-specific depending on the problem we are trying to solve.<br><br>

- **Classification** task → given the Data Scientist's dating history → predict the likelihood of a new "person of interest" of saying "yes" or "no" to the date
    - Ground Truth / label / target / dependent variable / y = `date` column<br><br>
- **Regression** task → given the Data Scientist's dating history → predict the most likely age of a new "person of interest" who says "yes"
    - Ground Truth / label / target / dependent variable / y = `age` column

## Evaluation Metrics

- How do we know whether the predictions made by our model is worth paying attention to?
    - We need to figure out how to evaluate them<br><br>
- **Regression Evaluation Metrics** → used to evaluate the performance of a regression model
    - Basic Idea: Calculate _how much_ the model predictions are off by, on average, wrt ground truth
    - Common metrics include
        - Mean squared error (MSE) 
        - Root mean squared error (RMSE)
        - Mean average error (MAE)
        - Mean average percent error (MAPE)<br><br>
- **Classification Evaluation Metrics** → used to evaluate the performance of a classification model
    - Basic Idea: Calculate _how many_ model predictions are off by, on average, wrt ground truth
    - In binary classification, we often use the terms _positive_ and _negative_ classes. By convention:
        - Positive class → presence of what we want to predict (e.g. said yes to the date, customer is loyal, tumour is cancerous)
        - Negative class → absence of what we want to predict (e.g. said no to the date, customer is disloyal, tumour is benign)
        - Positive/Negative distinction also dependent on the task we are trying to solve:
            - For eg, loyalty model → loyal customer is the positive class
            - customer churn model → disloyal customer is the positive class<br><br>
    - Common metrics include 
        - **Accuracy:** measures how many data points correctly predicted (true positives & true negatives)<br><br>
        - **Precision:** measures false positives against true positives
            - When cost of misclassifying a negative class is high (eg website blocker for kids should not misclassify adult content as "safe for kids")
            - Not as much a problem if kid-friendly website is misclassified as "unsafe for kids"<br><br>
        - **Recall:** measures false negatives against true positives
            - When cost of misclassifying a positive class is high (eg cancer detection algo should not misclassify cancerous tumours as benign)
            - False diagnosis of a benign tumour not so much a problem<br><br>
        - **F1-Score:** Combine Precision & Recall, s.t. class imbalance is managed and a trade-off between the 2 metrics is achieved
            - Consider the case of the website blocker for adult content
            - Classify every website as adult content → No False Positives (High Precision), Many False Negatives (Low Recall)
            - Classify every website as kid-friendly content → Many False Negatives (Low Precision), No False Positives (High Recall)
            - Often times, realistically, we need to trade one metric off with the other<br><br>
        - **ROC-AUC:** Another often recommended metric, it helps evaluate the performance of the classification model at different prediction cut-offs (or thresholds)
            - Basic idea: Classification models actually calculate likehood of a data point belonging to a particular class
            - Likelihood lies between 0 and 1 → think of it as extent of trust that model places in its decision
            - Typically, if likelihood of positive class > 0.5, data point classified as positive
            - 0.5 is referred to as prediction cut-off or threshold → this is a general number and not based on some scientific fact
            - ROC-AUC evaluates the performance of the classification model at different predictions thresholds<br><br>
    - We often use a **confusion matrix** to visually showcase the model performance
        - Sometimes, ground truth is along the vertical axis whereas predictions are along horizontal

## Bias and Variance

- When training a model, we have a few things to keep in mind
- We want the model to be flexible enough to learn from the data, but not so much that it learns erroneous patterns as well
    - In ML, this idea is characterised by what is called the **Bias vs Variance tradeoff**

<figure>
  <img src='https://www.cs.cornell.edu/courses/cs4780/2018fa/lectures/images/bias_variance/bullseye.png' style='width:400px'/>
</figure>

- Suppose you are putting together a professional shooting team (or your post-apocalypse zombie-hunting band of survivors!)
    - Figure above shows target shooting results of 4 shooters<br><br>
- **Bias** → how far the shots are from the bullseye, on average
    - On average, shots fired by shooters with low bias are _not very far from the center_
    - High bias → on average, shots are _farther away from the center_<br><br>
- **Variance** → how much your results differ from each other, on average (spread of the area where your shots land)
    - Low variance means that if you fire multiple times, on average there isn’t that much difference in where you hit
    - High variance → on average, your shots are all over the place<br><br>
- There are 4 kinds of shooters here:
    1. _Low Bias, Low Variance_ → _Perfect Fit_ → Always going to hit the bullseye
        - Hit the bullseye → low bias
        - Always → low variance
        - You definitely want this person in your shooting team<br><br>
    2. _Low Bias, High Variance_ → _Underfitting_ → On average, will not be far from the target, but individual shots may hit quite far from the target
        - Not too far from target → low bias
        - Individual shots dispersed → high variance
        - Not too reliable but can do something brilliant, once in a while
        - Your focus → reduce the spread of the shots<br><br>
    3. _High Bias, Low Variance_ → _Overfitting_ → Does not do "exactly" what you might hope for but is reliable and consistent
        - Shots farther from target → high bias
        - Less dispersed spread → low variance
        - Might not be as effective as you might want but you have a pretty good idea of what the shooter can achieve
        - Your focus → shift the center of the spread to the bullseye<br><br>
    4. _High Bias, High Variance_ → _Horrible Fit_ → Neither reliable nor accurate - all over the place except for the bullseye
        - Shots too far from target → high bias
        - Shots too dispersed → high variance
        - You definitely don't want this person in your shooting team<br><br>
- The different shooters are the different ML algorithms 
    - The shots are the predictions that algos make for different data points!        

### Overfitting and Underfitting

- Goal: End up with the "Just right" model!

- Models with high bias → too oversimplified to be flexible and learn the actual underlying complex relationships
    - With low variance, it is commonly referred to as an "underfit" model
    - Subpar performance in both training and test data
    - Easier problem to solve → opt for a more complex model<br><br>
- Models with high variance → too inconsistent to be robust and demonstrate comparable performance in real world unseen data
    - With low bias, it is commonly referred to as an "overfit" model
    - Drastic difference in performance between training and test data
    - Slightly more difficult to solve → slowly reduce model complexity (if reduced quickly, you will end up underfitting)<br><br>
- There are specific strategies and approaches we use to manage the bias-variance tradeoff → will be covered in later notebooks

## Summary of Learning Outcomes

1. [Machine Learning](#ml)
     - Machine Learning leverages the power of computers to apply statistical principles to build intelligent systems that analyze and learn from massive amounts of data.
     - "Machine Learning algorithms enable the computers to learn from data, and even improve themselves, without being explicitly programmed."
<br><br>

2. [Categories of Machine Learning](#cat)
    - Supervised Learning has the "Ground Truth" or answers an ML algorithm uses to learn and predict for new data.
    - Unsupervised Learning is used to explore patterns, trends, and find the inherent structure of the data. 
    - Reinforcement Learning is not as common and uses a "rewards-based" algorithm
<br><br>

3. [Supervised Learning](#sup)
    - Supervised Learning has either Regression or Classification algorithms
        - Regression models predict continuous values
        - Classification models predict discrete values
<br><br>

4. [Evaluation Metrics](#em)
    - Both Regression and Classification have multiple ways to test the accuracy of a model
<br><br>

5. [Bias and Variance](#bias)
    - Tradeoff between Bias and Variance in a model
    - Too much bias and the model is too simple. It did not learn enough from the training data.
    - Too much variance and the model is too complex. It learned too much from the training data that it does not perform well to unseen or new data. This is also known as Overfitting.
