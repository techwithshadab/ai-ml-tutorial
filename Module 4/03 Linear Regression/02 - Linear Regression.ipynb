## Linear Regression
### What is Linear Regression?

* A regression line is the line that best "fits" the data points. 
    - If you look at the figure below, you can even imagine a line yourself that would go through the middle of all those points.  
* In machine learning, we can use that line to predict future data points   
* Predict a real value known as a the Dependent Variable (DV, y, label, or target) from a set of Independent Variables (IV or features)   
* Linear line is used to model the relationship between IV and DV   
* Can be used for forecasting   
* Great model to use when data shows linear trend

#### Dreaded Formula

- For all you who took math in high school → must have learnt the line equation as `y = mx + b`
    - `y` is the _dependent_ variable
    - `x` is the _independent_ variable
    - `b` is the _y intercept_ which is the value y when x = 0
    - `m` is the _slope_ → rate at which `y` changes with `x` → we increase x by 1, y changes by a factor of m
- In ML and statistics, we have a slightly different formula which is $\hat{Y_i} = b_1 X_1 + b_2 X_2 + ... + b_n X_n + b_0$ where $n$ is the total number of attributes/features/columns
    - Don't be afraid! It's very similar:
    1. $\hat{Y_i}$ is the "predicted" value of our y
        - Just think of it as a guess/estimate of "y"
    2. $X_i$ is a feature/attribute where `i` is represents one feature out of a set of features
        - Think of it as one or more `x`s, that's it!
        - Say you have features like transaction amount, purchase, age of customer etc. Then, $X_1$ is transaction amount, $X_2$ is purchase etc
    3. $b_0$ is the y-intercept where x = 0, i.e., where the line crosses the x axis
        - Regardless of the number of features you have → only 1 $b_0$
    4. $b_i$ is the coefficient for feature $X_i$ x (equivalent of the slope (or "m") for that feature)
        - Remember that each feature (x) has it's own "b"
- `b`s are referred to as **model coefficients**
    - To create your model, it must "learn" good values for these coefficients from the data that you have
    - Once you have good `b` values, the model is ready to predict `y` values for new data points that it hasn't even seen before
- Coefficients are estimated using the **least squares criterion** → a mathematically sound algorithm to find the line (aka get the values of these coefficients that define the line) that minimizes the sum of squared errors (or residuals). In other words:
    - Reduce the error between our prediction and the actual values
    - This method leads to the best fitted **regression line** that fits the data points

- TLDR:
    - It all boils down to `y = mx + b`
    - Reducing the errors in prediction is how we get our regression line
- We'll explore this more as we go

Let's look at a simple example below:

<figure>
  <img src='img/linreg_form_ex_1.png' width='75%'/>
  <figcaption>Image from Launchpad.AI</figcaption>
</figure>

- Above is an example of a Linear Regression model with a single feature/attribute, $X$
- Look at the equation in the middle `y^ = 2.46x + 4.89` → the numbers are the **model coefficients**
    - $b_0$ → 4.89, $b_1$ → 2.46
- This is how we get our predictions
- For e.g., let's look at the first line:
    - `y^ = b1x + b0`
    - `y^ = 2.46x + 4.89`
    - `y^ = 2.46 * (9) + 4.89`
    - `y^ = 27.04`
- Then, we compare `y^ = 27.04` to `y = 27.75`
    - First number is prediction and second number is the true/actual value
    - Difference between them → Error/Residual → `y - y^` or `27.75 - 27.04` = `0.71`
    - Squared Error/Squared Residual → `(y - y^)^2` or `0.71^2` = `0.01`
    - The term above is used within the mathematical algorithm to influence how good model coefficients are learned and arrived at

### Evaluation Metric

- How do we judge if a model is performing well?   
    - We do so via evaluation metrics
- They capture the performance of a model on the task of house price prediction
  
**Linear Regression Evaluation Metrics.**

- One common metric is called **R Squared**
  - Percentage variation in `y`, explained by all the x variables together
  - Its value is (usually) between 0 and 1 and it indicates the strength of Linear Regression model
- Higher the R2 value, the less scattered the data points which ultimately means a good model. Lower R2 value means the data points are more scattered and the model is not good.
      
<figure>
  <img src='img/linreg_r2_1.png' width='50%'/>
  <figcaption>Image from <a href='https://www.qualtrics.com/support/stats-iq/analyses/regression-guides/user-friendly-guide-linear-regression/' style='width:500px'>Qualtrics.com</a></figcaption>
</figure>

- Another metric is the **MSE** or _Mean (of) Squared Error_
    - It accounts for the error between the predicted value (regression line) and the actual value.
        - In our example above, the MSE is derived from the difference between the average price of the home (the line) and the actual data point.
    - The closer the number to "0" is better.
        - Intuitively, we want a smaller error in our prediction
- Let's look at what our MSE would be using just the average price of homes:

````python
# Here's some math to get MSE
df0['Error'] = df0.PRICE.apply(lambda y: y - y_bar)
df0['Error_sq'] = df0.Error.apply(lambda e: e**2)

sse = np.sum(df0.Error_sq)
mse = np.mean(df0.Error_sq)
print('The mean squared error (MSE) is %f' % mse)
````

- The MSE is about 84.4. 
    - In and of itself, the value is meaningless.
    - However, it is a good benchmark to judge against future models

## Feature Selection: Intro to Correlation

- Numerous methods to perform feature selection on a dataset exist
    - They will be covered in the rest of the notebook
- For now, let's focus on **Correlation**
- Correlation measures the strength of the linear relationship between two variables/attributes/features
- A typical example is height and weight (or say, income and taxes paid). Those two variables are often correlated. Why is this so?

<figure>
  <img src='img/corr.png'/>
  <figcaption>Image from <a href='https://www.simplypsychology.org/correlation.html'>SimplyPsychology: Correlation</a></figcaption>
</figure>

**What does this mean for Machine Learning?**

- Correlation is **good** for a feature (x) used to predict a target variable (y)
- However, there can be issues if there is correlation BETWEEN features.  
- This phenomenon is known as **Multicollinearity**
    - When the two independent variables are *highly* correlated that there becomes a linear association between the two
- For e.g. consider a small set of features used to measure university acceptance like
    1. Student's overall grades
    2. GPA
    3. SAT Score
    4. Extracurricular activities
- GPA has a high correlation with the student's overall grades. 
    - So much so in fact that if the grades flunctuate in any way, it could be seen that the GPA would fluctuate with it
    - Fluctuations in grade → same fluctuation in GPA → chances of an amplified fluctuation in predicted variable
    - What we want: independent signals!  
- Sometimes, accuracy may not be affected. But, the results might not be **interpretabile** is important.
- Interpretability just means the ability to explain the "ins-and-outs" of how you came to your prediction.
    - Some algorithms are very difficult to explain to stakeholders.
    - We'll see examples of these during the course

````python
#set the graph style
sns.set(style="white")

# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=bool) # Return an array of zeros with the same shape and type as a given array
mask[np.triu_indices_from(mask)] = True # Return the indices for the upper-triangle of arr

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
````

- In the cell above, the mask is used to hide the upper triangle of the heat map.

    - More **blue** the square is, the more negatively correlated the variables are

- Many types of scaling, available in sklearn, of which 2 most common ones are:
    - **Standardization**
    - **Min-Max Scaling**
- Standardization/Z-Score (sklearn: `StandardScaler`)
    - It rescales the features to have the properties for a standard normal distribution
        - mean of 0 and standard deviation of 1
    - Works quite well if data is close to being normally distributed
- Min-Max Scaling (sklearn: `MinMaxScaler`)
    - Changes range of the data to be between 0 and 1
    - Min is pegged to 0 and max is pegged to 1, basicallyy
    - Useful when data is not normally distributed

## Data Leakage

- Let's revisit Bias and Variance
- QUESTION: What's wrong with training our models on ALL of our data? 

### train_test_split

- ML "learns" from the data it is exposed to 
    - No matter the algorithm, if it data is exposed to it (during fitting, transformation etc), it will be used
    - Control the exposure → ultimately, we want a ML model that performs in the real world, on data that it is yet to see
- Often times, we have one single big dataset to use for models
- Divide them into "training" and a "test" set. 
- Model only "sees" the training data when it learns → i.e., it is not exposed to the test data
  - Prevents data leakage
- "Test" set is only used to finally evaluate the model

<figure>
  <img src='img/train_test_split.png' width='50%'/>
  <figcaption>Image from <a href='https://towardsdatascience.com/train-test-split-and-cross-validation-in-python-80b61beca4b6'>TowardsDataScience: train-test-split and cross validation</a></figcaption>
</figure>

- Let's use it with our dataset:

````python
#import train_test_split
from sklearn.model_selection import train_test_split

#denote our features. We'll use the unscaled features from our original DataFrame
X = df.drop(columns=['PRICE']) 
X.head()
````

- Below is the function for train_test_split()<br><br>

````python
#Splitting our dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
````

- Inside the function we have four arguments:
    - Our feature variables (X)
    - Our dependent variable (y)
    - test_size (0.33) which means how much of the dataset will be split into the test set
        - 1/3 for test set and 2/3 for training data set
    - random_state (42) is any number used to reproduce the same results. 
        - It remembers the random split of our data.
- Look at the **left** side of the equation:
    - We have our training features set (X_train)
    - Testing features set (X_test)
    - Training target set (y_train)
    - Testing target set (y_test)

What happens after using `train_test_split`?
- Original data is split with 67% as data for training and 33% to be tested
- Can verify this by looking at the shapes

````python
# Verifies if condition is true
assert X_train.shape[0] + X_test.shape[0] == X.shape[0]
X_train.shape, y_train.shape, X_test.shape, y_test.shape
````

- Let's use our split dataset for prediction:

````python
# Initialize ML algorithm
lr = LinearRegression()

# Train our model with our training set
lr.fit(X_train, y_train) #Formely (X, y)

# Predict the price and save it to a variable
y_pred = lr.predict(X_test) #Formely X

# Plot
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred, edgecolors=(0, 0, 0))
ax.plot([y_test.min(), y_test.max()], 
        [y_test.min(), y_test.max()], 
        'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()

# compute MSE
mse_split = mean_squared_error(y_test, y_pred)
print('The mean squared error (MSE) is %f' % mse_split)
````

- The MSE is very similar to our MSE trained on our full dataset.
- However, our model is now more **robust** and should perform better with new or unseen data we collect

````python
#peek at our dataset
df.head()
````

````python
#Save the features with correlation >= (+/-) 0.3
X = df.drop(columns=["PRICE", "CHAS","DIS"])
X.head()
````

- We split and then scale our data to prevent data leakage
- Method adapted from this [article](https://datascience.stackexchange.com/questions/54908/data-normalization-before-or-after-train-test-split).

````python
# Split our data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Initialize Scaler 
scaler = StandardScaler()

# fit and transform our training features
# Remember, we don't need to scale our dependent variable
X_train_scaled = scaler.fit_transform(X_train)

# Initialize ML algorithm
lr = LinearRegression()

# Fit our data to the model
lr.fit(X_train_scaled, y_train)

# Scale our test features based on the parameters learned from our training features
# Note the difference of .transform() method instead of .fit_transform() like before
X_test_scaled = scaler.transform(X_test)

# Predict the price and save it to a variable
y_pred = lr.predict(X_test_scaled)

# Plot
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred, edgecolors=(0, 0, 0))
ax.plot([y_test.min(), y_test.max()], 
        [y_test.min(), y_test.max()], 
        'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()

# compute MSE
mse_fin = mean_squared_error(y_test, y_pred)
print('The mean squared error (MSE) is %f' % mse_fin)
````
