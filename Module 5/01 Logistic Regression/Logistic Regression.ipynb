### Intro to Logistic Regression

- Similar to Linear Regression, Logistic Regression is a popular, quick, and simple ML algorithm
- Like Linear Regression, Logistic Regression uses a set of independent variables (x) to predict an outcome (y)
      
<details>
<summary>Click here for a visual linear vs logistic comparison</summary>

<figure>
  <img src='img/linlog.png' width='60%'/>

</details>

- Linear Regression → predict a continuous value (ex. weight)
- Logistic Regression has only two possible outcomes for y: 0 or 1
    - Often as "classes"
- Examples of this include:
    1. a phone-text classifier to tell us if an incoming text should be marked "spam" or "not spam"
    2. radiological images to predict whether a tumor is benign or malignant
    3. Review credit history to predict whether a customer will default on their loan repayments or repay the loan
    4. Predicting if an anonymous data scientist will get a date... (yes or no)
- Logistic Regression to predict membership to a certain category
- A few business use-cases include:
    1. Qualify Leads: Used to segment users into distincit categories for business intelligence
        - ex. Which Dockers email subscribers will convert into buyers
        - With this information, one could streamline sales operations to focus on prospects who have the higher likelihood of buying
    2. Recommend Products (especially useful for digital businesses)
    - For instance, calculate the likelihood (probability) of a customer purchasing a brand new line of Levi's jeans
        - Range from 0 to 1 for 0 to 100% chance of occurence
    - If the likelihood of a customer buying these new jeans is greater than a certain threshold (for example, 80%), recommend the product for upselling on the website.
    - This increases overall sales because we're recommending products that are more likely to be purchased
- Advantages of Logistic Regression for deployment
    1. Ease of use
        - Like Linear Regression, training and maintaining the model has less engineering overheard than more complex models
    2. Interpretability
        - Logistic regression is relatively straightforward to interpret
        - Model can help us assess which input variable is responsible for the greatest change in predicted value
    3. Scalability
        - Algorithm is very efficient
        - Trains faster than other models
        - Easier to scale with a larger volume increase
    4. Real-time predictions
        - Because of the low engineering overhead, model can be retrained and generate predictions in near real-time

### How Do We Classify?

- We have the same issue as we did with Linear Regression. How do we classify these points?
- Logistic Regression uses what's called a **Decision Boundary**
    - It's a threshold that's used to categorize the probabilities into discrete classes. For e.g. if we set 0.5 as a threshold
    - `y = 0 if predicted probability < 0.5`
        - A customer is not predicted to default if the probability of them defaulting is less than 50%<br><br>
    - `y = 1 if predicted probability is > 0.5`
        - A customer is predicted to default if the probability of them defaulting is greater than 50%

- In a nutshell, it shows our model's prediction vs the actual value for each class
- This is what it looks like for our data:

```python
#importing sklearn confusion matrix
from sklearn.metrics import confusion_matrix
    
# Creating a confusion matrix
con_mat = confusion_matrix(y_test, pred)
con_mat = pd.DataFrame(con_mat, range(2), range(2))
   
#Ploting the confusion matrix
plt.figure(figsize=(6,6))
plt.title("0 = No Churn          1 = Churn") # The spacing is intentional
sns.set(font_scale=1.5) 
sns.heatmap(con_mat, annot=True, annot_kws={"size": 16}, fmt='g', cmap='Blues', cbar=False)
plt.ylabel("Actual Values")
plt.xlabel("Predicted Values")
plt.show()    
```

```python
# YOUR TURN: sklear also has a function ConfusionMatrixDisplay to plot and visualize the confusion matrix
# Check it out! 

```

1. The **diagonal** represents the number of correct classifications for each class: 
    - Look at the top left square:
        - Class 0: 1130/1234 were correctly predicted
        - 1130 represents the term **True Negatives (TN)**
        - Of all the 1234 total customers who **did not** churn, our model correctly predicted 1130 of them.
        - This represents the number of predictions that were correctly predicted to be class 0 (no churn)
    - Look at the bottom right square:
        - This is our class of interest
        - Class 1: 224/524 were correctly predicted
        - 224 represents the term **True Positives (TP)**
        - Of all the 524 total customers who **did** churn, our model correctly predicted 224 of them. 
        - This represents the number of predictions that were correctly predicted to be class 1 (churn)
2. Therefore the other elements, of the matrix, are the misclassified values: 
    - Look at the top right square:
        - Class 0: 104/1234 were incorrectly predicted
        - 104 represents the term **False Positives (FP)**
        - Of all the 1234 total customers who **did not** churn, our model misclassified 104 of them. 
        - This represents the number of predictions that were incorrectly predicted to be class 0 (churn)
    - Look at the bottom left square:
        - Class 1: 300/524 were incorrectly predicted
        - 300 represents the term **False Negatives (FN)**
        - Of all the 524 total customers who **did** churn, our model misclassified 300 of them.
        - This represents the number of predictions that were incorrectly predicted to be class 0 (no churn)
- Another way to visualize the Confusion Matrix is via the Pandas `.crosstab()` method
- Note: Crosstab allows us to see the **supports** of the matrix which are just the totals
    - Represented by the `All` row and column 
    - Example: Total (1430) of all the customers (1130 + 300) the model **predicted** to **not** churn.

```python
# Pandas crosstab method creates a confusion matrix for us
df_matrix = pd.crosstab(y_test, pred, rownames=['True'], colnames=['Predicted'], margins=True)
df_matrix
```

- The Confusion Matrix is the basis for all the other classification evaluation metrics
  - It's with this raw data (TP, TN, FP, FN) that the other metrics are calculated
  - Remember our `77%` accuracy score?
      - That too was calculated from this matrix

<a id="f1"></a>
#### F1 Score

- How do we score our model given that we have both metrics Precision and Recall?
- This is where the F1 score comes in → "harmonic mean" between precision and recall
    - A form of weighted average of both
    - It's a good metric to use when both precision and recall are important
-  What’s the difference between this metric and accuracy?
    - False positives (FP) and false negatives (FN) can really hurt a model, while true negatives (TN) are often less important to what we're trying to solve especially in a business setting.
    - The F1 score tries to take this into account, giving more weight to false negatives (FN) and false positives (FP) while not letting large numbers of true negatives (TN) influence your score.
        - We can see that our f1-score of 0.53 paints a different portrait than our original accuracy score of 0.77 would lead us to believe
    - For our **imbalanced** dataset, we would care more about the F1 score for our **Churned (1)** class than Not Churned **(0)**.   
- F1 Score is a good starting metric for evaluating binary and imbalanced classification

<a id="acc"></a>
#### Accuracy, Macro Avg, Weighted Avg

- **Accuracy** is simply the amount of correct predictions over our total.
    - It's easy to miss in the Classification Report, and it's the third number under the "f1-score" column
    - It's a more useful metric when both classes are more balanced and are equally important in the study
- **macro avg** is simply the unweighted average of both classes for each metric in its column
    - For example, In the "precision" column, 0.79 + 0.68 = 0.74
    - It does the same for recall and f1-score
- **weighted avg** does some fancy math to add weights to the class' average
    - Essentially, it adds a weight depending on the number of **True** labels for the class
    - Therefore, it tends to favor the majority class
        - Not as useful for imbalanced classification where we care more about the minority class
- Note: **macro avg and weighted avg** are good for seeing an overall picture of the scores
    - Used more for **multi-class** classification (ours is just binary)

### ROC-AUC

- Chances are we will never get perfect accuracy between both precision and recall
- In fact, there is often a "tug of war" between both metrics

<figure>
  <img src='img/pr_tug.png'/>
  <figcaption>Image from <a href='https://medium.com/geekculture/accuracy-precision-and-recall-in-plain-english-1c90c92a6c88'>Medium</a></figcaption>
</figure>

- In binary classification, we often make trade-offs between higher precision or higher recall
- Optimizing for one hurts the other which leads to more errors:
  - False Positivies or False Negatives
- In our dataset, we're trying to figure out which customers will churn (class 1) out of all the customers. For the customers who we predict to churn, we will offer them a discount to hopefully prevent them from leaving
- Our model correctly predicts 224 churned customers out of 524 total. 
    - This is 224/524 which equals 0.43 which is our Recall.
    - 43% is pretty low
- It costs the company more money to get new customers than to keep the ones we have. However, if we want our model to catch more churned customers then the trade-off will be incorrectly classifing customers we predicted to churn but in fact did not leave
    - In the Confusion Matrix, these are known as False Positives (FP)
    - We could tune the model to get 100% of the churned customers, but then we would have too many FPs
    - In the business context, it would be the equivalent of offering discounts to everyone which would be a waste of money for those who were never planning on leaving            
- Isn't there a way we could see the differing Recall percentages against the amount of False Positives? In fact there is!

- **ROC Curve - AUC Score** (Receiver Operating Characteristics curve (ROC) and Area under the Curve (AUC) Score) are 2 very popular metrics among data scientists.
- Let’s start understanding this with an example:
- We have a classification model that gives probability values ranging between 0–1 to predict the probability of a customer churning or not.
- A probability score near 0 indicates a very low probability that the customer will churn whereas a probability near 1 indicates a very high probability of a customer churning.
- Now, by default if we consider a threshold of 0.5 then all the customers assigned probabilities ≤0.5 will be classified as “No Churn” and people assigned probabilities >0.5 will be classified as “Churn”.
    - This is what we did when we predicted using our Logistic Regression model
    - Each person was assigned a probability, and based on that number, was labeled as either "churn" or "no churn"
    - Sklearn did it automatically so we never actually saw the probabilities for each customer
- The ROC curve shows us how the model predicted the classes (churn or not churn) if the threshold was different
    - Instead of the default threshold 0.50, the ROC shows what happens if we vary it     
    - For example, What if we make it 0.3 or 0.9?

<a id="roc"></a>

#### Plot the ROC

- A few characteristics of the AUC for a ROC curve:
    - Only concerned about the score for the positive (1) class
    - It is a popular diagnostic tool for classifiers on balanced and imbalanced binary prediction problems alike because it is not biased to the majority or minority class.
- ROC-AUC consists of:
    - **False Positive Rate (FPR, fall-out)** on the x-axis
        - This is the number of falsely predicted values in 1 class (0s predicted as 1s) with respect to all 0 data points
    - Mathematically, $\frac{FP}{FP + TN}$    
    - **True Positive Rate (TPR, recall)** on its y-axis
        - This is the number of correctly predicted values in 1 class (1s predicted as 1s) with respect to all 1 data points
        - Mathematically, $\frac{TP}{TP + FN}$
- It might make more sense once we implement it via sklearn
    - Note, we implement the ROC-AUC after the step where we train our Logistic Regression model
        - Usually, we would predict `x_test` to get the class label. Instead, we use the `predict_proba()` method to get the probabilities.
- Take a look at it below

```python
pred_proba = logreg.predict_proba(x_test)
df_pred_proba = pd.DataFrame(pred_proba * 100, columns=['% Not Churn', '% Churn'])
df_pred_proba.head()
```

- Above, each row is a customer with the model's predicted probability they will either churn or not churn
- We turned the array into a DataFrame to help with understanding, but here is what the actual array looks like:

```python
# array of probability predictions
pred_proba
```

- The resulting output is an array of lists where each list consist of 2 numbers. 
    - All probabilities on the left side is the probability of a data point being classified as 0 (no churn).
    - The right is the probability for 1 (churn).
- We can now take this list of probabilities and use it to calculate the ROC curve 
    - Again we are only focusing on class 1 (or the right side) which is all values with index

```python
from sklearn.metrics import roc_curve, auc

# Note that the third return value is not needed but still needs a placeholder "_"
# fpr = False Positive Rate, tpr = True Positive Rate 
fpr, tpr, _ = roc_curve(y_test, pred_proba[:, 1]) #pred_proba[:,1] just means the 2nd column (class 1)

fpr[100], tpr[100]
```

- You can see that the resulting output for fpr and tpr is an array.
- If we choose a random index 100, we get two values.
    - This corresponds to a coordinate on the ROC curve.

```python
# Compute AUC score
roc_auc = auc(fpr, tpr)

# Plot ROC curve with AUC score
import matplotlib.pyplot as plt

plt.figure()
lw = 2
plt.plot(fpr, tpr, color='b', lw=lw, 
         label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='r', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

**Let's look at the graph:**

- The `fpr` (False Positive Rate) was plotted along the x-axis
- The `tpr` (True Positive Rate) was plotted along the y-axis
- Each of the points of the (fpr, tpr) pair are plotted on the plain which leads to the blue line
- We can think of the plot as the fraction of correct predictions for the positive class (y-axis) versus the fraction of errors for the negative class (x-axis).
- The **red line** is a benchmark used to gauge the curve
    - It indicates the points where tpr = fpr
    - It's the equivalent of a random classifier or a 50-50 coin toss, and any point of the curve above red line means the model is better than random chance
- Pay attention to how the **blue line** curves.
    - On the top right, tpr = 1 and fpr = 1
        - This means at this threshold, the model would correctly classify 100% of the customers who churned but also incorrectly predict customers as churning, when they in fact did not, 100% of the time
    - On the bottom left is the other extreme which is tpr = 0 and fpr = 0
        - At this point, our model incorrectly classifies churning customers as non-churning 100% of the time, and correctly classifies non-churning customers as non-churning 100% of the time
            - Note: equation for fpr: 
                - FPR = 1 - TNR (or True Negative Rate)
                - Substitute fpr = 0 in the equation: 0 = 1 - TNR
                - TNR = 1 (or 100%)
    - Either extreme is not helpful to us
- Why is the curve useful?
    - The curve is useful because we have visual of the tradeoffs between `fpr` and `tpr`
    - For example, let's say your boss wants our model to be able to correctly predict churning customers at least 80% of the time, on average. 
        - If you go to `0.8` on the y-axis, you'll see that it corresponds to about `0.25` on the x-axis
        - This means that in order for our model to correctly predict churning customers an average of at least 80% of the time, we'll have to be willing to accept that it also misclassifies non-churning customers as churning about 25% of the time

<a id="auc"></a>

#### AUC

- Another reason why the ROC curve is useful is that we can get the Area Under the Curve (AUC) which is a useful metric for scoring and comparing models
    - Ranges from 0 (very poor) to 1 (very good)
- There are two ways to compute the AUC score through Sklearn:
    1. `roc_curve()` -> `auc()`: Used for plotting the AUC
    2. `roc_auc_score()`: When only trying to see the score
- First, compute the AUC score with `auc()`
    - We actually already did this when we plotted our ROC curve
    - We'll put the code below again

```python
#import roc_curve and the auc functions
from sklearn.metrics import roc_curve, auc

# compute the fpr and tpr 
fpr, tpr, _ = roc_curve(y_test, pred_proba[:, 1])

# Compute AUC score
roc_auc = auc(fpr, tpr)
print("ROC-AUC Score:", roc_auc)
```

- Note: The ROC-AUC score can only be predicted using the **probabilities** of the prediction. For classifiers in sklearn, use `predict_proba()` instead of `predict`.
- The other way we can calculate the ROC-AUC score is directly via the `roc_auc_score()` as seen below:

```python
from sklearn.metrics import roc_auc_score

# Predict with y_test and pred_proba
roc_auc = roc_auc_score(y_test, pred_proba[:, 1])

print(f'ROC-AUC Score: {roc_auc}')
```

- Below is the **incorrect** way to calculate auc score:<br><br>

```python
# WRONG
roc_auc_wrong = roc_auc_score(y_test, pred)

print(f'Wrong ROC-AUC Score: {roc_auc_wrong}')
```

- This is what happens when you use the wrong one. The ROC-AUC score is wrong.
- ALWAYS use `predict_proba()` not `predict()` for the ROC-AUC
- Although very popular, the ROC-AUC is not perfect
    - For imbalanced classification with a severe skew and few examples of the minority class, the ROC AUC can be misleading.
    - This is because a small number of correct or incorrect predictions can result in a large change in the ROC Curve or ROC AUC score.
- A common alternative is the precision-recall curve and area under curve (PR-AUC).
